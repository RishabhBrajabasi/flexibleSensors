```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'hand'
df = pd.read_csv(filename+'.csv', header=None)
fig = plt.figure()
time_col = df[0]
lux_col = df[1]
plt.plot(time_col, lux_col)
fig.suptitle(filename, fontsize=20)
plt.xlabel('time', fontsize=16)
plt.ylabel('lux', fontsize=16)
fig.savefig(filename)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filenames = ['Z1-gold-paint','Z1-gold','Z1-plain','Z4-clear','Z4-ClearShrink-40mm','Z4-EmbeddedShrink-40mm','Z4-paint']
dfs = []
for filename in filenames:
    dataframe = pd.read_csv('raw/Z-series tests/coat/'+filename+'.csv', header=None)
    dfs.append(dataframe)
```

```python
dfs[0].head()
```

```python
try:
    time_col = dfs[0][0]
    lux_col = dfs[0][1]
    plt.plot(lux_col,time_col)
except KeyError:
    lux_col = dfs[0][0]
    plt.plot(lux_col)
```

```python
def plot_graph(index):
    fig = plt.figure()
    try:
        time_col = dfs[0][0]
        lux_col = dfs[0][1]
        plt.plot(time_col,lux_col)
    except KeyError:
        lux_col = dfs[0][0]
        plt.plot(lux_col)
    finally:
        fig.suptitle(filenames[index], fontsize=20)
        plt.xlabel('time', fontsize=16)
        plt.ylabel('lux', fontsize=16)
        fig.savefig('Results/Z1_4-CoatingTest'+filenames[index])
        
```

```python
for i in range(0,len(dfs)):
    plot_graph(i)
```

```python
fig = plt.figure()
for i in range(3,len(dfs)):
    try:
        time_col = dfs[i][0]
        lux_col = dfs[i][1]
        plt.plot(time_col,lux_col)
    except KeyError:
        lux_col = dfs[i][0]
        plt.plot(lux_col)
    fig.suptitle('Z4-All Three Coating', fontsize=20)
    plt.xlabel('time', fontsize=16)
    plt.ylabel('lux', fontsize=16)
    plt.savefig('Z4-AllThreeCoating')
```
